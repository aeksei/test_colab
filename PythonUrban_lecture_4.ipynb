### Скачивание файлов из локальной системы

`files.upload` возвращает словарь загруженных файлов.  
Словарь задается именем файла, а значениями являются данные,  
которые были загружены.

````python
from google.colab import files

uploaded = files.upload()

for file_name, file in uploaded.items():
  print(f'Файл "{file_name}" с длиной {len(file)} байт')
  print(file)
````

````python
for file_name, file in uploaded.items():
  uploaded[file_name] = file.decode()  # сделать из bytes -> str
````

````python
import json

filename = "2021_12_28_Archive_2021_City_chronic_with_themes.geojson"
json.loads(uploaded[filename])
````

### Файлы и менеджер контекста `with`
Python по умолчанию достаточно просто работает  
с файлами операционной системы.  
Перед работой с файлом надо его открыть  
с помощью команды `open`.
```python
f = open('path/to/file', 'filemode', encoding='utf8')
```

Результатом этой операции будет указатель на файл.

Давайте по порядку разберем все аргументы:
- `path/to/file` путь к файлу может быть относительным или абсолютным.
- `filemode` режим, в котором файл нужно открывать.  
Записывается в виде строки, состоит из следующих букв:
 - **r** – открыть на чтение (по умолчанию)
 - **w** – перезаписать и открыть на запись (если файла нет, то он создастся)
 - **x** – создать и открыть на запись (если уже есть – исключение)
 - **a** – открыть на дозапись (указатель будет поставлен в конец)
 - **t** – открыть в текстовом виде (по умолчанию)
 - **b** – открыть в бинарном виде
- `encoding` – указание, в какой кодировке файл записан (utf8, cp1251 и т.д.)

При открытии файла в режимах на запись  
ставится блокировка на уровне операционной системы.

После работы с файлом его нужно закрыть.  
Особенно это важно, когда мы начинаем работать  
с файлами на запись.  

```python
f = open(...)  # открываем файл
...  # работаем с файлом
f.close()  # закрываем файл
```

Если не закрыть – будет висеть, пока не закроете Python или  
сборщик мусора (garbage collector) не решит удалить его.

Для явного указания места работы с файлом, а также чтобы не забывать  
закрыть файл после обработки, существует менеджер контекста `with`.  

В блоке менеджера контекста открытый файл «жив» и с ним можно работать,  
при выходе из блока – файл закрывается.

Блок `with` сокращает запись при работе с блоком try-except
```python
try:
    f = open(...)  # открываем файл
    ...  # работаем с файлом
except Exception:  # вдруг возникнет какая-то ошибка
    ...
finally:
    f.close()  # файл закроется в любом случае и не потеряем частично записанные данные
```

```python
with open(...) as f:  # открываем файл
    ...  # работаем с файлом
```


````python
import json


tmp_file = os.path.join(tmp_dir, "temp.json")
with open(tmp_file, "w") as f:  # записали содержимое во временный файл
    json.dump({"simple_json": 123}, f, indent=4)
````

````python
with open(tmp_file, "r") as f:  # открыли файл для чтения
    print(json.load(f))  # с помощью модуля json считали содержимое
````

### Удаление файлов и папок

Для того чтобы удалить файл есть функция `os.remove()`.  

````python
os.remove(tmp_file)
os.path.exists(tmp_file)  # проверяем существует ли файл
````

А вот функция `os.rmdir()` удаляет директорию но только пустую.

````python
os.rmdir(tmp_dir)
os.path.exists(tmp_dir)  # проверяем существует ли папка
````
